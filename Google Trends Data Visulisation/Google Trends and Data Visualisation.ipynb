### Convert Strings to DateTime Objects

### Converting from Daily to Monthly Data

[Pandas .resample() documentation](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.resample.html) <br>

```python
#To convert our daily data into monthly data, we're going to use the .resample() function
#The only things we need to specify is which column to use (i.e., our DATE column) and what kind of sample frequency we want (i.e., the "rule"). 
#We want a monthly frequency, so we use 'M'.

#the average price over the course of the month
df_btc_monthly = df_btc_price.resample('M', on='DATE').mean()

df_btc_monthly = df_btc_price.resample('M', on='DATE').last() #price of the month - the price at month-end.

print(df_btc_monthly.shape)
print(df_btc_monthly.head())
```

```python
#Using Locators and DateFormatters to generate Tick Marks on a Time Line
#we need a YearLocator() and a MonthLocator() objects, which will help Matplotlib find the years and the months. 
#Then we also need a DateFormatter()

years = mdates.YearLocator()
months = mdates.MonthLocator()
years_fmt = mdates.DateFormatter("%Y")
```

```python
#Data Visualisation - Tesla Line Charts in Matplotlib
#Increases size and resolution
plt.figure(figsize=(14,8), dpi=120) 
plt.title('Tesla Web Search vs Price', fontsize=18)

#Plot the Tesla stock price against the Tesla search volume using a line chart and two different axes. 
ax1 = plt.gca()    # get current axes
ax2 = ax1.twinx()  # share  the same x_axis

#add some labels,colour, and fontsize in the axes
ax1.set_ylabel('TSLA Stock Price',color='#E6232E',fontsize=14)
ax2.set_ylabel('Search Trend',color='skyblue',fontsize=14)

#with rotation added to the x-ticks. With .set_ylim() and .set_xlim() 
#you have precise control over which data you want to show on the chart
#Set the minimum and maximum values on the axes
ax1.set_ylim([0, 600])
ax1.set_xlim([df_tesla.MONTH.min(), df_tesla.MONTH.max()])

# format where the major and minor ticks should be using the Locators.
ax1.xaxis.set_major_locator(years)
ax1.xaxis.set_major_formatter(years_fmt)
ax1.xaxis.set_minor_locator(months)

#add linewidth for larger charts
ax1.plot(df_tesla.MONTH,df_tesla.TSLA_USD_CLOSE,color='#E6232E',linewidth=3)
ax2.plot(df_tesla.MONTH,df_tesla.TSLA_WEB_SEARCH,color='skyblue',linewidth=3)

#Displays chart explicitly
plt.show()
```
